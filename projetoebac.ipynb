# Análise de Dados de Crédito

Nesta análise, exploraremos um conjunto de dados que inclui informações sobre clientes de uma instituição financeira. O objetivo é entender o comportamento dos clientes em relação ao crédito, transações e perfis demográficos. A partir dessa análise, podemos gerar insights úteis para a tomada de decisões.

## Passos:
1. Exploração dos dados
2. Manipulação e limpeza dos dados
3. Visualizações
4. Insights e Storytelling

# Importação de bibliotecas e carregamento dos dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


  
df = pd.read_csv('/kaggle/input/credito-csv/credito.csv', na_values='na')


df.head()
```

# Limpeza de dados

```python
def convert_to_float(value):
    try:
        
        value = value.replace('$', '').replace('.', '').replace(',', '.')
        return float(value)
    except:
        return None


df['limite_credito'] = df['limite_credito'].apply(convert_to_float)
df['valor_transacoes_12m'] = df['valor_transacoes_12m'].apply(convert_to_float)


df[['limite_credito', 'valor_transacoes_12m']].head()

```

# Visualizações

```python
df.replace([float('inf'), -float('inf')], pd.NA, inplace=True)

sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.histplot(df['idade'], kde=True, bins=20, color='skyblue')
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(x='tipo_cartao', y='limite_credito', data=df, palette='Set2')
plt.title('Limite de Crédito por Tipo de Cartão')
plt.xlabel('Tipo de Cartão')
plt.ylabel('Limite de Crédito')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(x='qtd_transacoes_12m', y='valor_transacoes_12m', data=df, color='coral')
plt.title('Relação entre Valor de Transações e Número de Transações')
plt.xlabel('Quantidade de Transações em 12 meses')
plt.ylabel('Valor das Transações em 12 meses')
plt.show()
```

## Conclusão e Insights

### 1. Perfil Demográfico
A distribuição da idade dos clientes revela que a maioria se concentra na faixa etária entre 30 e 50 anos, com um pico em torno de 40 anos.

### 2. Limite de Crédito e Tipo de Cartão
Clientes com cartões de categorias superiores têm limites de crédito significativamente mais altos.

### 3. Relação entre Valor e Quantidade de Transações
Existe uma correlação positiva entre o número de transações e o valor total das transações, indicando que clientes mais ativos movimentam maiores quantias.

Esses insights podem ser usados para estratégias de segmentação e gestão de riscos.
